```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox
from experta import *

class CarProblem(Fact): #كلاسات لتعريف الفاكت يعني الأعطال الأساسية
    pass

class DiagnosticFact(Fact): #كلاسات لتعريف فاكت جديدة يعني الحقائق لي بستنتجها وبرجع بستخدما
    pass

class ToyotaCorolla2015(KnowledgeEngine):
    @Rule(CarProblem(problem='engine_wont_start'))
    def check_battery(self):
        self.suggest("Check the battery and starter motor.", 'battery_issue') # عرفنا فاكت جديدة لأعطال البطارية

    @Rule(DiagnosticFact(newfact='battery_issue'))
    def suggest_battery_maintenance(self):
        self.suggest("Battery issue detected. Check the battery connections and charge.")

    @Rule(CarProblem(problem='clicking_sound'))
    def check_battery_clicking(self):
        self.ask_question("Is there a clicking sound when trying to start the engine?", 'clicking_sound_issue') #فاكت لصوت تكتكة

    @Rule(DiagnosticFact(newfact='clicking_sound_issue'))
    def suggest_clicking_sound_solution(self):
        self.suggest("Clicking sound detected. Check the battery connections and starter motor.")

    @Rule(CarProblem(problem='squealing_noise'))
    def check_brake_pads(self):
        self.suggest("Inspect the brake pads and discs for wear.", 'brake_system_issue') #فاكت للفريمات

    @Rule(CarProblem(problem='brake_warning_light'))
    def check_brake_system(self):
        self.suggest("Brake warning light is on. Inspect brake pads and fluid level.", 'brake_system_issue')

    @Rule(DiagnosticFact(newfact='brake_system_issue'))
    def suggest_brake_service(self):
        self.suggest("Brake system issue detected. Consider a full brake service.")

    @Rule(CarProblem(problem='trouble_shifting_gears'))
    def check_transmission(self):
        self.suggest("Inspect the transmission fluid level and condition. Consider a transmission service.")

    @Rule(CarProblem(problem='transmission_warning_light'))
    def transmission_warning(self):
        self.suggest("Transmission warning light is on. Have the transmission system checked by a professional.")

    @Rule(CarProblem(problem='engine_overheating'))
    def check_coolant(self):
        self.suggest("Check the coolant level and inspect the radiator for leaks.", 'cooling_system_issue') #فاكت لنظام التبريد

    @Rule(CarProblem(problem='steam_from_hood'))
    def overheating_steam(self):
        self.suggest("Engine is overheating. Check the cooling system immediately.", 'cooling_system_issue')

    @Rule(DiagnosticFact(newfact='cooling_system_issue'))
    def suggest_cooling_system_service(self):
        self.suggest("Cooling system issue detected. Service the cooling system.")

    @Rule(CarProblem(problem='uneven_tire_wear'))
    def check_suspension(self):
        self.suggest("Inspect the suspension system including shocks and struts.", 'suspension_issue') #فاكت لهيكل السيارة

    @Rule(CarProblem(problem='car_pulls_to_one_side'))
    def check_alignment(self):
        self.suggest("Car alignment may be off. Check the wheel alignment.", 'alignment_issue') #فاكت لدوزان الدواليب

    @Rule(DiagnosticFact(newfact='suspension_issue'))
    def suggest_suspension_service(self):
        self.suggest("Suspension issue detected. Service the suspension system.")

    @Rule(DiagnosticFact(newfact='alignment_issue'))
    def suggest_alignment_service(self):
        self.suggest("Alignment issue detected. Perform a wheel alignment.")

    @Rule(CarProblem(problem='dim_lights'))
    def check_alternator(self):
        self.suggest("Check the alternator and battery connections.", 'electrical_issue') #فاكت لمشاكل الكهربا

    @Rule(CarProblem(problem='malfunctioning_accessories'))
    def check_electrical_system(self):
        self.suggest("Inspect the car's electrical system for faults.", 'electrical_issue')

    @Rule(DiagnosticFact(newfact='electrical_issue'))
    def suggest_electrical_service(self):
        self.suggest("Electrical issue detected. Service the car's electrical system.")

    @Rule(CarProblem(problem='loud_exhaust_noise'))
    def check_exhaust_system(self):
        self.suggest("Check the exhaust system for leaks or damage.", 'exhaust_issue') #فاكت للأشيكمان

    @Rule(CarProblem(problem='check_engine_light'))
    def check_engine(self):
        self.suggest("Check engine light is on. Perform a diagnostic scan to identify the issue.", 'engine_issue') #فاكت للمحرك

    @Rule(CarProblem(problem='poor_fuel_efficiency'))
    def check_fuel_system(self):
        self.suggest("Inspect the fuel system including fuel filter and injectors.", 'fuel_system_issue') #فاكت للفيول

    @Rule(DiagnosticFact(newfact='exhaust_issue'))
    def suggest_exhaust_service(self):
        self.suggest("Exhaust system issue detected. Service the exhaust system.")

    @Rule(DiagnosticFact(newfact='engine_issue'))
    def suggest_engine_service(self):
        self.suggest("Engine issue detected. Service the engine.")

    @Rule(DiagnosticFact(newfact='fuel_system_issue'))
    def suggest_fuel_system_service(self):
        self.suggest("Fuel system issue detected. Service the fuel system.")

    def suggest(self, message, newfact=None): #ليطبع المسج لليوزر وإذا لزم نعمل فاكت جديدة
        print(message)
        if newfact:
            self.declare(DiagnosticFact(newfact=newfact))

    def ask_question(self, question, newfact=None): #لنسأل اليوزر لناخد قرارات أبعد وبعدها ممكن نعرف فاكت جديدة
        answer = input(f"{question} (yes/no): ").lower()
        if answer == 'yes':
            self.declare(DiagnosticFact(newfact=newfact))
        elif answer == 'no':
            pass  

class CarDiagnostic: #
    def __init__(self, master):
        self.master = master
        master.title("Toyotal Corolla 2015 Diagnostic System")

        self.label = tk.Label(master, text="Select a problem:")
        self.label.pack(pady=10)

        self.problem_options = [
            "Engine won't start",
            "Clicking sound",
            "Squealing noise",
            "Brake warning light",
            "Trouble shifting gears",
            "Transmission warning light",
            "Engine overheating",
            "Steam from hood",
            "Uneven tire wear",
            "Car pulls to one side",
            "Dim lights",
            "Malfunctioning accessories",
            "Loud exhaust noise",
            "Check engine light",
            "Poor fuel efficiency"
        ]

        self.problem_selected = tk.StringVar(master)
        self.problem_selected.set(self.problem_options[0])  

        self.problem_dropdown = ttk.Combobox(master, textvariable=self.problem_selected, values=self.problem_options, width=50)
        self.problem_dropdown.pack(pady=5)

        self.diagnose_button = tk.Button(master, text="Diagnose", command=self.diagnose_problem)
        self.diagnose_button.pack(pady=10)

    def diagnose_problem(self):
        selected_problem = self.problem_selected.get()

        engine = ToyotaCorolla2015()
        engine.reset()

        if selected_problem == "Engine won't start":
            engine.declare(CarProblem(problem='engine_wont_start'))
        elif selected_problem == "Clicking sound":
            engine.declare(CarProblem(problem='clicking_sound'))
        elif selected_problem == "Squealing noise":
            engine.declare(CarProblem(problem='squealing_noise'))
        elif selected_problem == "Brake warning light":
            engine.declare(CarProblem(problem='brake_warning_light'))
        elif selected_problem == "Trouble shifting gears":
            engine.declare(CarProblem(problem='trouble_shifting_gears'))
        elif selected_problem == "Transmission warning light":
            engine.declare(CarProblem(problem='transmission_warning_light'))
        elif selected_problem == "Engine overheating":
            engine.declare(CarProblem(problem='engine_overheating'))
        elif selected_problem == "Steam from hood":
            engine.declare(CarProblem(problem='steam_from_hood'))
        elif selected_problem == "Uneven tire wear":
            engine.declare(CarProblem(problem='uneven_tire_wear'))
        elif selected_problem == "Car pulls to one side":
            engine.declare(CarProblem(problem='car_pulls_to_one_side'))
        elif selected_problem == "Dim lights":
            engine.declare(CarProblem(problem='dim_lights'))
        elif selected_problem == "Malfunctioning accessories":
            engine.declare(CarProblem(problem='malfunctioning_accessories'))
        elif selected_problem == "Loud exhaust noise":
            engine.declare(CarProblem(problem='loud_exhaust_noise'))
        elif selected_problem == "Check engine light":
            engine.declare(CarProblem(problem='check_engine_light'))
        elif selected_problem == "Poor fuel efficiency":
            engine.declare(CarProblem(problem='poor_fuel_efficiency'))

        engine.run()

        messagebox.showinfo("Diagnostic Result", "Diagnostic completed. Check console for details.")

if __name__ == "__main__":
    root = tk.Tk()
    app = CarDiagnostic(root)
    root.mainloop()
```
